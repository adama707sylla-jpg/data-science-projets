```python
import numpy as np
import pandas as pd
import requests as r
```

```python
user_url = "https://dummyjson.com/users"
response = r.get(user_url, timeout = 10)
data = response.json()
df_user = pd.DataFrame(data['users'])
df_user.head()
```

```python
cart_url = "https://dummyjson.com/carts"
response = r.get(cart_url,timeout = 10)
data = response.json()
df_cart = pd.DataFrame(data['carts'])
df_cart.head()
```

```python
df_cart.columns
```

```python
product_url = "https://dummyjson.com/products"
response = r.get(product_url, timeout=10)
data = response.json()
df_product = pd.DataFrame(data['products'])
df_product.head()
```

```python
#extraire les infos d'adresse pour des colonnes propre
df_adrees = pd.json_normalize(df_user['address'])
df_user = pd.concat([df_user.drop('address', axis=1), df_adrees], axis = 1)
df_user[['firstName', 'lastName','city','state']].head()
```

```python
df_user.head()
```

```python
df_crypto = pd.json_normalize(df_user['crypto'])
df_user = pd.concat([df_user.drop(columns=['crypto']), df_crypto], axis = 1)
df_user[['firstName','lastName','coin','wallet']]
```

```python
df_user.head()
```

```python
#nettoyage pour extraire les donnees dans la colonne products
#ici on va utiliser les deux dataframes products et carts

#premierement on explose la colonne products
df_cart_exp = df_cart.explode(['products'])

#apres on normalize cad transformer la dict en colonne
df_product_info = pd.json_normalize(df_cart_exp['products'])

#on reinintialise les index
df_cart_exp = df_cart_exp.reset_index(drop=True)
df_product_info = df_product_info.reset_index(drop=True)

#on concatene le tout en supprimant l'ancienne colonne products
df_cart_final = pd.concat([df_cart_exp.drop(columns=['products']), df_product_info], axis = 1)
df_cart_final.head()
```

```python
#calcule du montant total pour chaque ligne
df_cart_final['amount'] = df_cart_final['price'] * df_cart_final['quantity']
df_cart_final['amount']
```
